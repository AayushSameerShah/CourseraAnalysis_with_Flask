```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.animation import PillowWriter
```

```python
%matplotlib notebook
```

```python
background= "#121f3b"
face = "#fffffa"
```

```python
#First1
fig, ax = plt.subplots(figsize= (5,2), facecolor= background)
ax.set_facecolor(background)
y = np.random.randn(1000).cumsum()
x = np.arange(len(y))

def animate(i):
    ax.clear()
    ax.set_xticks([])
    ax.set_yticks([])
    for side in ['left', 'right', 'top', 'bottom']:
        ax.spines[side].set_visible(False)
    ax.plot(x[i:i+50], y[i:i+50], color= face, lw= 3)
    

anim = FuncAnimation(fig,
                     animate,
                     frames= np.arange(0, 1000),
                     interval= 500,
                     repeat= False)

writergif = PillowWriter(fps=60) 
anim.save('./webpage/static/sample1-black.gif', writer=writergif)
```

```python
data = np.random.normal(size= 500)
```

```python
#Second2
fig, ax = plt.subplots(figsize= (5,6), facecolor= background)
ax.set_facecolor(background)

def animate(i):
    ax.clear()
    ax.set_xticks([])
    ax.set_yticks([])
    for side in ['left', 'right', 'top', 'bottom']:
        ax.spines[side].set_visible(False)
    plt.hist(data[:i], bins= 15, color= face)
    
anim = FuncAnimation(fig,
                     animate,
                     frames= np.arange(0, 500),
                     interval= 100,
                     repeat= False)

writergif = PillowWriter(fps=60) 
anim.save('./webpage/static/sample2-black.gif', writer=writergif)
```
